```python
#importing packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
# setting train path
train_path = "images/Train/"
```

```python
# getting class names
traincounts = Counter(os.listdir(train_path))
```

```python
# Plot the class distribution
plt.figure(figsize=(10, 5))
sns.barplot(x=list(traincounts.keys()), y=list(traincounts.values()), palette="viridis")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.title("Image Count per Class")
plt.xticks(rotation=45)  # Rotate class names if needed
plt.show()
```
